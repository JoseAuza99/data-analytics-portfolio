```python
#Proporcionar una funcin elaborada por ustedes para calcular la varianza. 
#Aplicar vuestra funcin sobre alguno de de los 3 pd.Series (edad, deporte o habilidad) y
#comprobar que obtenis los mismos resultados que utilizando la funcin np.var.

def formula_varianza(vector):
    n= len(vector)
    cuadrados= 0
    media= np.mean(vector)
    for numero in vector:
        cuadrados= cuadrados + (numero-media)**2
    resultado = cuadrados/(n-1)
    return resultado

if formula_varianza(edad) == np.var(edad, ddof=1):
    print(f'Para la serie edad la formula calcula correctamente la varianza, que es igual a: {formula_varianza(edad):.2f}')

```

R\ Para la funcion vamos a sumar los cuadrados de la diferencia de cada elemento con la media y al final dividir eso entre n-1 ya que es una varianza muestral. Al final vemos que obtebemos lo mismo que con la funcion de numpy

```python
# Tendencia central (modo, media y mediana)
#Posicin (cuartiles, percentiles y deciles)
#Dispersin (rango, rango intercuartlico, varianza y desviacin tpica)
#Forma (coeficiente de asimetra de Fisher "skewness" y coeficiente de curtosis "kurtosis"])

def summary(df):
    shape= df.shape
    columnas= df.columns
    diccionario= {}
    varibles_categoricas= []
    variables_con_nulos= []
    for i in range(shape[1]):
        if df[columnas[i]].dtype == int or df[columnas[i]].dtype == float:
            if df[columnas[i]].isnull().any():
                variables_con_nulos.append(columnas[i])
            media= np.mean(df[columnas[i]])
            mediana= np.median(df[columnas[i]])
            moda= st.mode(df[columnas[i]])
            Q1= np.quantile(df[columnas[i]],0.25)
            Q2= np.quantile(df[columnas[i]],0.5)
            Q3= np.quantile(df[columnas[i]],0.75)
            Q4= np.quantile(df[columnas[i]],1)
            varianza= np.var(df[columnas[i]], ddof=1)
            std= np.std(df[columnas[i]], ddof=1)
            diccionario[columnas[i]]= [media,mediana,moda,Q1,Q2,Q3,Q4,varianza,std]
        else:
            varibles_categoricas.append(columnas[i])
    Data_frame= pd.DataFrame(diccionario, index=['media','mediana','moda','Q1','Q2','Q3','Q4','varianza','std'])
    return print(f'Variables categoricas: {varibles_categoricas}', f'\nEstas variables tienen valores nulos: {variables_con_nulos}\n',
                f'\nResumen variables numericas:\n {Data_frame}')

summary(estudiantes)

       
            
    

```
