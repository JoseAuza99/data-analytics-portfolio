#### Analizamos que es cada campo:


- startDate → Fecha y hora en que se inició la medición o el registro de datos.
- endDate → Fecha y hora en que finalizó la medición o el registro de datos.
- creationDate → Fecha en la que el dato fue registrado en la base de datos (puede diferir de las fechas anteriores si los datos fueron importados de otro dispositivo o sincronizados después).

- value → El valor registrado en la medición (ejemplo: pasos, frecuencia cardíaca, calorías quemadas, etc.).
- unit → La unidad de medida correspondiente al valor (value). Puede ser "count", "bpm" (latidos por minuto), "kcal", "mg/dL", etc.
- type → El tipo de medición realizada (ejemplo: "HKQuantityTypeIdentifierStepCount" para conteo de pasos, "HKQuantityTypeIdentifierHeartRate" para frecuencia cardíaca).

- sourceName → El nombre de la fuente que registró los datos (puede ser el iPhone, un Apple Watch u otra app como Strava, MyFitnessPal, etc.).
- sourceVersion → La versión del sistema operativo o app que generó los datos.
- device → Información sobre el dispositivo que capturó los datos (modelo, versión del sistema operativo, etc.).

#### Metadatos en el dataframe:

- meta_HKTimeZone → Zona horaria en la que se registró el dato (puede estar vacío si no se capturó).
- meta_HKWasUserEntered → Indica si el usuario ingresó manualmente el dato (1.0 si fue ingresado, vacío si fue medido automáticamente).
- meta_HKAlgorithmVersion → Número de versión del algoritmo de Apple usado para calcular ciertos datos (ejemplo: VO2 máx, calorías activas).
- meta_HKDateOfEarliestDataUsedForEstimate → Fecha de los datos más antiguos usados en estimaciones de salud (puede aplicarse a métricas calculadas como VO2 Máx).
- ##### meta_HKMetadataKeyHeartRateMotionContext → Contexto en el que se midió la frecuencia cardíaca (ejemplo: si fue en reposo, durante ejercicio o en movimiento).
- meta_HKMetadataKeySyncVersion → Versión de sincronización de datos entre dispositivos (Apple Watch, iPhone).
- meta_HKMetadataKeyAppleDeviceCalibrated → Indica si el dispositivo estaba calibrado para ciertas mediciones (por ejemplo, distancia recorrida o calorías quemadas).
- meta_HKVO2MaxTestType → Tipo de prueba utilizada para estimar el VO2 Máx. (rendimiento cardiovascular).
- ##### meta_HKMetadataKeyDevicePlacementSide → Indica en qué lado del cuerpo estaba colocado el dispositivo (ejemplo: muñeca izquierda o derecha en el Apple Watch).
- meta_HKMetadataKeySyncIdentifier → Identificador único de sincronización de datos entre dispositivos.

 - Teniendo en cuenta los datos analizados observamos que para metadatos el señalado en negrita meta_HKMetadataKeyHeartRateMotionContext puede ser interesante, el resto de metadatos, podriamos decir q no interesan en este analisis.

- Adicionalmente por cantidad de registros podriamos guardar meta_HKMetadataKeyDevicePlacementSide → Indica en qué lado del cuerpo estaba colocado el dispositivo (ejemplo: muñeca izquierda o derecha en el Apple Watch), ya que tiene pocos nulos en comparación al resto de meta datos y podria ser relativamente interesante el dato.

🔥 Gasto Energético
- HKQuantityTypeIdentifierBasalEnergyBurned → Energía quemada en reposo (calorías que tu cuerpo usa para funciones básicas como respirar y mantener la temperatura).
- HKQuantityTypeIdentifierActiveEnergyBurned → Energía quemada durante la actividad física.

🏃 Actividad Física y Movimiento
- HKQuantityTypeIdentifierDistanceWalkingRunning → Distancia recorrida caminando o corriendo (en metros o kilómetros).
- HKQuantityTypeIdentifierStepCount → Conteo de pasos diarios.
- HKQuantityTypeIdentifierWalkingStepLength → Longitud promedio de los pasos al caminar (en metros).
- HKQuantityTypeIdentifierWalkingSpeed → Velocidad promedio al caminar (metros por segundo).
- HKQuantityTypeIdentifierWalkingDoubleSupportPercentage → Porcentaje del tiempo que ambos pies están en contacto con el suelo mientras caminas (indica estabilidad).
- HKQuantityTypeIdentifierWalkingAsymmetryPercentage → Diferencia en la duración de cada paso al caminar (puede indicar problemas de equilibrio o movilidad).
- HKQuantityTypeIdentifierFlightsClimbed → Número de pisos subidos (medido por altímetro).

❤️ Frecuencia Cardíaca y Respiratoria
- HKQuantityTypeIdentifierHeartRate → Frecuencia cardíaca (latidos por minuto, BPM).
- HKQuantityTypeIdentifierRestingHeartRate → Frecuencia cardíaca en reposo.
- HKQuantityTypeIdentifierWalkingHeartRateAverage → Frecuencia cardíaca promedio mientras caminas.
- HKQuantityTypeIdentifierHeartRateVariabilitySDNN → Variabilidad de la frecuencia cardíaca (indicador de salud cardiovascular y estrés).
- HKQuantityTypeIdentifierVO2Max → Máximo consumo de oxígeno (indicador de capacidad aeróbica).
- HKQuantityTypeIdentifierRespiratoryRate → Frecuencia respiratoria (respiraciones por minuto).

🎧 Exposición al Ruido
- HKQuantityTypeIdentifierHeadphoneAudioExposure → Nivel de exposición al audio en auriculares (decibeles, dB).
  
��️ Ejercicio y Postura
- HKQuantityTypeIdentifierAppleExerciseTime → Minutos de actividad física intensa.
- HKQuantityTypeIdentifierAppleStandTime → Minutos de pie durante el día.
- HKCategoryTypeIdentifierAppleStandHour → Indica si estuviste de pie al menos 1 minuto en cada hora.

😴 Sueño
- HKCategoryTypeIdentifierSleepAnalysis → Análisis del sueño (duración, fases, etc.).
- HKDataTypeSleepDurationGoal → Meta de duración del sueño establecida.

🚴 Otros Datos de Salud
- HKQuantityTypeIdentifierBodyMass → Peso corporal (kg).
- HKQuantityTypeIdentifierHeight → Altura (metros o cm).
- HKQuantityTypeIdentifierDietaryWater → Cantidad de agua ingerida (ml).
- HKQuantityTypeIdentifierDistanceCycling → Distancia recorrida en bicicleta (km o metros).
- HKQuantityTypeIdentifierSixMinuteWalkTestDistance → Distancia recorrida en una prueba de caminata de 6 minutos (métrica usada en pruebas médicas).

```python
# reemplazaremos los valores de tipo por nombres mas comunes y 
#agregaremos una columna tipo de actividad para que las divida segun corresponda
# Diccionario de mapeo: 'tipo original' -> ('nuevo nombre', 'tipo de actividad')
type_mapping = {
    # 🔥 Gasto Energético
    "HKQuantityTypeIdentifierBasalEnergyBurned": ("energia_quemada_reposo", "Gasto Energético"),
    "HKQuantityTypeIdentifierActiveEnergyBurned": ("energia_quemada_activa", "Gasto Energético"),

    # 🏃 Actividad Física y Movimiento
    "HKQuantityTypeIdentifierDistanceWalkingRunning": ("distancia_caminata_carrera", "Actividad Física"),
    "HKQuantityTypeIdentifierStepCount": ("conteo_pasos", "Actividad Física"),
    "HKQuantityTypeIdentifierWalkingStepLength": ("longitud_paso", "Actividad Física"),
    "HKQuantityTypeIdentifierWalkingSpeed": ("velocidad_caminata", "Actividad Física"),
    "HKQuantityTypeIdentifierWalkingDoubleSupportPercentage": ("porcentaje_doble_apoyo", "Actividad Física"),
    "HKQuantityTypeIdentifierWalkingAsymmetryPercentage": ("porcentaje_asimetria_caminata", "Actividad Física"),
    "HKQuantityTypeIdentifierFlightsClimbed": ("pisos_subidos", "Actividad Física"),

    # ❤️ Frecuencia Cardíaca y Respiratoria
    "HKQuantityTypeIdentifierHeartRate": ("frecuencia_cardiaca", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierRestingHeartRate": ("frecuencia_cardiaca_reposo", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierWalkingHeartRateAverage": ("frecuencia_cardiaca_caminata", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierHeartRateVariabilitySDNN": ("variabilidad_frecuencia_cardiaca", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierVO2Max": ("vo2max", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierRespiratoryRate": ("frecuencia_respiratoria", "Salud Cardiovascular"),

    # 🎧 Exposición al Ruido
    "HKQuantityTypeIdentifierHeadphoneAudioExposure": ("exposicion_audio", "Exposición al Ruido"),

    # 🏋️ Ejercicio y Postura
    "HKQuantityTypeIdentifierAppleExerciseTime": ("tiempo_ejercicio", "Ejercicio y Postura"),
    "HKQuantityTypeIdentifierAppleStandTime": ("tiempo_de_pie", "Ejercicio y Postura"),
    "HKCategoryTypeIdentifierAppleStandHour": ("hora_de_pie", "Ejercicio y Postura"),

    # 😴 Sueño
    "HKCategoryTypeIdentifierSleepAnalysis": ("analisis_sueno", "Sueño"),
    "HKDataTypeSleepDurationGoal": ("meta_duracion_sueno", "Sueño"),

    # 🚴 Otros Datos de Salud
    "HKQuantityTypeIdentifierBodyMass": ("peso_corporal", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierHeight": ("altura", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierDietaryWater": ("agua_consumida", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierDistanceCycling": ("distancia_ciclismo", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierSixMinuteWalkTestDistance": ("distancia_test_6min", "Otros Datos de Salud"),
}

# Crear las nuevas columnas con los valores mapeados
df_record["type"] = df_record["type"].map(lambda x: type_mapping[x][0] if x in type_mapping else x)
df_record["tipo_actividad"] = df_record["type"].map(lambda x: type_mapping[x][1] if x in type_mapping else "Desconocido")


```

```python
# Crear un diccionario basado en los nuevos nombres para asignar tipo de actividad
tipo_actividad_mapping = {v[0]: v[1] for v in type_mapping.values()}

# Asignar tipo de actividad basado en los nuevos valores de la columna "type"
df_record["tipo_actividad"] = df_record["type"].map(lambda x: tipo_actividad_mapping.get(x, "Desconocido"))

# Verificar cambios
print(df_record[["type", "tipo_actividad"]].head(10))

```
