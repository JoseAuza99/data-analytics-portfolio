#### Analizamos que es cada campo:


- startDate ‚Üí Fecha y hora en que se inici√≥ la medici√≥n o el registro de datos.
- endDate ‚Üí Fecha y hora en que finaliz√≥ la medici√≥n o el registro de datos.
- creationDate ‚Üí Fecha en la que el dato fue registrado en la base de datos (puede diferir de las fechas anteriores si los datos fueron importados de otro dispositivo o sincronizados despu√©s).

- value ‚Üí El valor registrado en la medici√≥n (ejemplo: pasos, frecuencia card√≠aca, calor√≠as quemadas, etc.).
- unit ‚Üí La unidad de medida correspondiente al valor (value). Puede ser "count", "bpm" (latidos por minuto), "kcal", "mg/dL", etc.
- type ‚Üí El tipo de medici√≥n realizada (ejemplo: "HKQuantityTypeIdentifierStepCount" para conteo de pasos, "HKQuantityTypeIdentifierHeartRate" para frecuencia card√≠aca).

- sourceName ‚Üí El nombre de la fuente que registr√≥ los datos (puede ser el iPhone, un Apple Watch u otra app como Strava, MyFitnessPal, etc.).
- sourceVersion ‚Üí La versi√≥n del sistema operativo o app que gener√≥ los datos.
- device ‚Üí Informaci√≥n sobre el dispositivo que captur√≥ los datos (modelo, versi√≥n del sistema operativo, etc.).

#### Metadatos en el dataframe:

- meta_HKTimeZone ‚Üí Zona horaria en la que se registr√≥ el dato (puede estar vac√≠o si no se captur√≥).
- meta_HKWasUserEntered ‚Üí Indica si el usuario ingres√≥ manualmente el dato (1.0 si fue ingresado, vac√≠o si fue medido autom√°ticamente).
- meta_HKAlgorithmVersion ‚Üí N√∫mero de versi√≥n del algoritmo de Apple usado para calcular ciertos datos (ejemplo: VO2 m√°x, calor√≠as activas).
- meta_HKDateOfEarliestDataUsedForEstimate ‚Üí Fecha de los datos m√°s antiguos usados en estimaciones de salud (puede aplicarse a m√©tricas calculadas como VO2 M√°x).
- ##### meta_HKMetadataKeyHeartRateMotionContext ‚Üí Contexto en el que se midi√≥ la frecuencia card√≠aca (ejemplo: si fue en reposo, durante ejercicio o en movimiento).
- meta_HKMetadataKeySyncVersion ‚Üí Versi√≥n de sincronizaci√≥n de datos entre dispositivos (Apple Watch, iPhone).
- meta_HKMetadataKeyAppleDeviceCalibrated ‚Üí Indica si el dispositivo estaba calibrado para ciertas mediciones (por ejemplo, distancia recorrida o calor√≠as quemadas).
- meta_HKVO2MaxTestType ‚Üí Tipo de prueba utilizada para estimar el VO2 M√°x. (rendimiento cardiovascular).
- ##### meta_HKMetadataKeyDevicePlacementSide ‚Üí Indica en qu√© lado del cuerpo estaba colocado el dispositivo (ejemplo: mu√±eca izquierda o derecha en el Apple Watch).
- meta_HKMetadataKeySyncIdentifier ‚Üí Identificador √∫nico de sincronizaci√≥n de datos entre dispositivos.

 - Teniendo en cuenta los datos analizados observamos que para metadatos el se√±alado en negrita meta_HKMetadataKeyHeartRateMotionContext puede ser interesante, el resto de metadatos, podriamos decir q no interesan en este analisis.

üéß Exposici√≥n al Ruido
- HKQuantityTypeIdentifierHeadphoneAudioExposure ‚Üí Nivel de exposici√≥n al audio en auriculares (decibeles, dB).
  
ÔøΩÔøΩÔ∏è Ejercicio y Postura

- HKQuantityTypeIdentifierAppleStandTime ‚Üí Minutos de pie durante el d√≠a.
- HKCategoryTypeIdentifierAppleStandHour ‚Üí Indica si estuviste de pie al menos 1 minuto en cada hora.

üò¥ Sue√±o
- HKCategoryTypeIdentifierSleepAnalysis ‚Üí An√°lisis del sue√±o (duraci√≥n, fases, etc.).
- HKDataTypeSleepDurationGoal ‚Üí Meta de duraci√≥n del sue√±o establecida.

üö¥ Otros Datos de Salud
- HKQuantityTypeIdentifierBodyMass ‚Üí Peso corporal (kg).
- HKQuantityTypeIdentifierHeight ‚Üí Altura (metros o cm).

```python
# reemplazaremos los valores de tipo por nombres mas comunes y 
#agregaremos una columna tipo de actividad para que las divida segun corresponda
# Diccionario de mapeo: 'tipo original' -> ('nuevo nombre', 'tipo de actividad')
type_mapping = {
    # üî• Gasto Energ√©tico
    "HKQuantityTypeIdentifierBasalEnergyBurned": ("energia_quemada_reposo", "Gasto Energ√©tico"),
    "HKQuantityTypeIdentifierActiveEnergyBurned": ("energia_quemada_activa", "Gasto Energ√©tico"),

    # üèÉ Actividad F√≠sica y Movimiento
    "HKQuantityTypeIdentifierDistanceWalkingRunning": ("distancia_caminata_carrera", "Actividad F√≠sica"),
    "HKQuantityTypeIdentifierStepCount": ("conteo_pasos", "Actividad F√≠sica"),
    "HKQuantityTypeIdentifierWalkingStepLength": ("longitud_paso", "Actividad F√≠sica"),
    "HKQuantityTypeIdentifierWalkingSpeed": ("velocidad_caminata", "Actividad F√≠sica"),
    "HKQuantityTypeIdentifierWalkingDoubleSupportPercentage": ("porcentaje_doble_apoyo", "Actividad F√≠sica"),
    "HKQuantityTypeIdentifierWalkingAsymmetryPercentage": ("porcentaje_asimetria_caminata", "Actividad F√≠sica"),
    "HKQuantityTypeIdentifierFlightsClimbed": ("pisos_subidos", "Actividad F√≠sica"),

    # ‚ù§Ô∏è Frecuencia Card√≠aca y Respiratoria
    "HKQuantityTypeIdentifierHeartRate": ("frecuencia_cardiaca", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierRestingHeartRate": ("frecuencia_cardiaca_reposo", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierWalkingHeartRateAverage": ("frecuencia_cardiaca_caminata", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierHeartRateVariabilitySDNN": ("variabilidad_frecuencia_cardiaca", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierVO2Max": ("vo2max", "Salud Cardiovascular"),
    "HKQuantityTypeIdentifierRespiratoryRate": ("frecuencia_respiratoria", "Salud Cardiovascular"),

    # üéß Exposici√≥n al Ruido
    "HKQuantityTypeIdentifierHeadphoneAudioExposure": ("exposicion_audio", "Exposici√≥n al Ruido"),

    # üèãÔ∏è Ejercicio y Postura
    "HKQuantityTypeIdentifierAppleExerciseTime": ("tiempo_ejercicio", "Ejercicio y Postura"),
    "HKQuantityTypeIdentifierAppleStandTime": ("tiempo_de_pie", "Ejercicio y Postura"),
    "HKCategoryTypeIdentifierAppleStandHour": ("hora_de_pie", "Ejercicio y Postura"),

    # üò¥ Sue√±o
    "HKCategoryTypeIdentifierSleepAnalysis": ("analisis_sueno", "Sue√±o"),
    "HKDataTypeSleepDurationGoal": ("meta_duracion_sueno", "Sue√±o"),

    # üö¥ Otros Datos de Salud
    "HKQuantityTypeIdentifierBodyMass": ("peso_corporal", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierHeight": ("altura", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierDietaryWater": ("agua_consumida", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierDistanceCycling": ("distancia_ciclismo", "Otros Datos de Salud"),
    "HKQuantityTypeIdentifierSixMinuteWalkTestDistance": ("distancia_test_6min", "Otros Datos de Salud"),
}

# Crear las nuevas columnas con los valores mapeados
df_record["type"] = df_record["type"].map(lambda x: type_mapping[x][0] if x in type_mapping else x)
df_record["tipo_actividad"] = df_record["type"].map(lambda x: type_mapping[x][1] if x in type_mapping else "Desconocido")


```

```python
# Crear un diccionario basado en los nuevos nombres para asignar tipo de actividad
tipo_actividad_mapping = {v[0]: v[1] for v in type_mapping.values()}

# Asignar tipo de actividad basado en los nuevos valores de la columna "type"
df_record["tipo_actividad"] = df_record["type"].map(lambda x: tipo_actividad_mapping.get(x, "Desconocido"))

# Verificar cambios
print(df_record[["type", "tipo_actividad"]].head(10))

```
